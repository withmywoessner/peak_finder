```python
df_temp = df_epoch.loc[(df_epoch['Channel'] != 'FCz') ]
#Rename subject to Subject
df_temp = df_temp.rename(columns={'subject': 'Subject'})

#Convert datatypes

df_ranges['Peak'] = df_ranges['Peak'].astype(int)
df_list = []
files = df_ranges['File'].to_numpy()
ranges = df_ranges['Range'].to_numpy()
peaks = df_ranges['Peak'].to_numpy()
#Convert peak array to int

for i in range(len(files)):
    files[i] = files[i] + "_" + str(peaks[i])
    #print(files[i])

#Convert numpy arrays to dictionary
dict_ranges = dict(zip(files, ranges))
dict_ranges
#Group by Subject and Channel 
df_no_cz_subject = df_temp.groupby(['Subject', 'Channel'])
df_list = []
for sub in df_no_cz_subject.groups:
    df_sub = df_no_cz_subject.get_group(sub)

    subject_num_df = df_sub['Subject'].iloc[0]
    print(subject_num_df)
    
    dict_key_gen = []
    for key in dict_ranges.keys():
        #print(key)
        dict_key_gen.append(key[:-2])

    if ((subject_num_df in dict_key_gen)):
        print("Subject found in ranges")
        subject_range_1 = dict_ranges[subject_num_df + "_1"]
        subject_range_2 = dict_ranges[subject_num_df + "_2"]
        subject_range_3 = dict_ranges[subject_num_df + "_3"]
    else :
        print("Subject not found in ranges")
        continue
    #Remove brackets and split string into list
    subject_range_1 = subject_range_1[1:-1]
    subject_range_1 = subject_range_1.split(', ')

    subject_range_2 = subject_range_2[1:-1]
    subject_range_2 = subject_range_2.split(', ')

    subject_range_3 = subject_range_3[1:-1]
    subject_range_3 = subject_range_3.split(', ')

    df_1 = df_sub.loc[(df_sub['time_ms'] >= float(subject_range_1[0])) & 
        (df_sub['time_ms'] <= float(subject_range_1[1]))]
    
    df_2 = df_sub.loc[(df_sub['time_ms'] >= float(subject_range_2[0])) &
        (df_sub['time_ms'] <= float(subject_range_2[1]))]
    
    df_3 = df_sub.loc[(df_sub['time_ms'] >= float(subject_range_3[0])) &
        (df_sub['time_ms'] <= float(subject_range_3[1]))]


    print('Find min and max')
    if df_1.empty or df_2.empty or df_3.empty:
        df_max1 = 0
        df_min2 = 0
        df_max3 = 0
        
    else: 
        df_max1 = df_1.loc[df_1['value'].idxmax()].copy()
        df_min2 = df_2.loc[df_2['value'].idxmin()].copy()
        df_max3 = df_3.loc[df_3['value'].idxmax()].copy()
    print('done')
    df_max1['Peak'] = '1'
    df_min2['Peak'] = '2'
    df_max3['Peak'] = '3'
    df_max1 = df_max1.to_frame().T
    df_min2 = df_min2.to_frame().T
    df_max3 = df_max3.to_frame().T
    df = pd.concat([df_max1, df_min2, df_max3])
    #print(df)
    df_list.append(df)



df_no_fcz = pd.concat(df_list)
```

```python
import datetime
now = datetime.datetime.now()
now = now.strftime("%Y-%m-%d_%H-%M")

df_no_fcz.to_csv('/home/woess/current assignment/ANT_redone/ANT14_redone_3_10/ANT14_long_no_fcz_' + now + '.csv', index=False)
```
