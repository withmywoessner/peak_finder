## Import packages and set up 

```python
import os
import pandas as pd
import numpy as np
df_list = []
path = "/home/woess/current assignment/ANT14_3_14/area"
#export_path = "/home/woess/current assignment/ANT_redone/ANT12"
os.chdir(path)

for filename in os.listdir(path):

    file = os.path.join(filename)
    if os.path.isfile(file) and file.endswith('.dat'):
        print(filename)
        # Read the csv file and append it to the list of data frames
        file_new = open(file)
        df_area = pd.read_csv(file_new, sep="\t")
        
        df_area.columns.values[2] = 'Area'
        df_area["Source File"] = filename
        last_underscore_loc = filename.rfind("_")
        df_area["area_num"] = filename[last_underscore_loc-3:last_underscore_loc]
        
        #Rename fourth column of dataframe
        
        df_list.append(df_area)

area_comb = pd.concat(df_list, ignore_index=True)
area_comb = area_comb.loc[area_comb["Source File"].str.contains("err")]
area_comb.head(1000)
```

```python
area_comb['File'] = area_comb['Source File'].str.rsplit('ern').str[0] + 'ERN_err'
#Drop the Source File column
area_comb = area_comb.drop(['Source File', 'Sweep Number'], axis=1)
#Rename Area to Ar
area_comb = area_comb.rename(columns={'Area': 'ar'})

#Rename ar1 to er0 in area_num column
area_comb['area_num'] = area_comb['area_num'].str.replace('ar1', 'er0')
#Rename ar2 to er1 in area_num column
area_comb['area_num'] = area_comb['area_num'].str.replace('ar2', 'er1')
#Rename ar3 to er2 in area_num column
area_comb['area_num'] = area_comb['area_num'].str.replace('ar3', 'er2')

#Sort by File
area_comb = area_comb.sort_values(by=['File', 'area_num'])
area_comb
```

```python
#Show duplicates
#Show all rows
pd.set_option('display.max_rows', 10)
#Remove duplicates
area_comb.drop_duplicates(inplace=True)
area_comb[area_comb.duplicated(keep=False)]
```

```python
wide_area_comb = area_comb.pivot(
                    index=['File'], 
                    columns=['Channel', "area_num"], 
                    values=["ar"]
                    )


value = wide_area_comb.columns.get_level_values(0)
channel = wide_area_comb.columns.get_level_values(1)
marker = wide_area_comb.columns.get_level_values(2)

# #Set the index to source_fileS
# wide_area_comb = wide_area_comb.set_index(('lab_num'))

#Rename the columns
wide_area_comb.columns = value + '_' + (marker) + "_" + channel + '_14'
```
